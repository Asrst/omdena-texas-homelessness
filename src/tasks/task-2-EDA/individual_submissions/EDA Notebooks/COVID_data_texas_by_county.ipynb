```python
df = pd.read_excel('omdena-texas-homelessness\src\data\health\COVID Related\Texas COVID-19 Case Count Data by County.xlsx', header=2)
df.dropna(axis = 0, inplace=True)
df1 = df.T
header_row = 0
df1.columns = df1.iloc[header_row]
df1 = df1.iloc[1:,:]

df1.index= df1.index.str.replace("Cases", "")
df1.index = pd.to_datetime(df1.index)
df1.groupby(by = [df1.index.month, df1.index.year])
df1 =df1[df1.Total != 0]
df1.columns = df1.columns.str.lower()

print(df1.columns, df1.index, df1.shape)
display(df1.head())
```
