This notebook has an initial Exploratory Data Analysis and Visualizations for the Data from Texas for the report Out of Reach 2021 (Full report in https://nlihc.org/sites/default/files/oor/2021/Out-of-Reach_2021.pdf). I extracted and clean the data for Texas as a csv file. 

Established in 1974 by Cushing N. Dolbeare, the National
Low Income Housing Coalition is dedicated solely to
achieving socially just public policy that ensures people with
the lowest incomes in the United States have affordable and
decent homes.
NLIHC provides up-to-date information, formulates
policy, and educates the public on housing needs and the
strategies for solutions. Permission to reprint portions of this
report or the data therein is granted, provided appropriate
credit is given to the National Low Income Housing
Coalition. Additional copies of Out of Reach are available
from NLIHC.

Additional local data can be found online at
www.nlihc.org/oor
The Print / PDF version of Out of Reach contains limited
data in an effort to present the most important information
in a limited number of pages.
The Out of Reach methodology was developed by Cushing
N. Dolbeare, founder of the National Low Income Housing
Coalition.
Data for other states, metropolitan areas, counties, and zip
codes can be found at http://nlihc.org/oor

Definitions:

FMR = Fiscal Year 2021 Fair Market Rent

AMI = Area Median Income

Affordable Rents represent the generally accepted standard of spending no more than 30% of gross income on rent and utilities.

```python
# There are 43 Metropolitan Areas and 254 Counties
```

```python
# Histograms for numerical values

df.hist(figsize=(20,15), bins=50)
plt.show()
```

```python
sns.scatterplot(x="Housing Costs: 2 BR FMR", y="Annual Area Median Income (AMI)", data=df)
plt.show()
```

20 Counties with **higher** number of full time jobs at mean renter wage needed (1.8 to 2.8) to afford 2 BR Fair Market Rent

```python
 
df.sort_values(by = "Full-time jobs at mean renter wage needed to afford 2 BR FMR", ascending = False).head(20)
```

20 Counties with **lower** number of full time jobs at mean renter wage needed (0.4 to 0.6) to afford 2 BR Fair Market Rent

```python
df.sort_values(by = "Full-time jobs at mean renter wage needed to afford 2 BR FMR", ascending = True).head(20)
```

**Housing Costs for renting 2BR vs Annual Area Median Income in both Metropolitan and Counties Area**

```python
sns.relplot(x="Housing Costs: 2 BR FMR", y="Annual Area Median Income (AMI)", data=df, kind="scatter", col="Area")
plt.show()
```

**Hourly wages necessary to afford a 2 BR vs Estimated hourly mean renter wage in Metropolitan Area and Counties**

```python
sns.relplot(x="Estimated hourly mean renter wage", y="Hourly wage necessary to afford 2 BR FMR", data=df, kind="scatter", col="Area")
plt.show()
```
