```python
#This simply does that if 50% of the column is NAN then we will drop it

df.dropna(axis=1, thresh = int(0.4*df.shape[0]), inplace=True) 
```

```python
df.describe()
```

```python
df.info()
```

```python
col = df.columns
```

```python
len(col)
```

```python
for i in range(len(col)):
    if df[col[i]].dtype == "int64":
        df[col[i]] = df[col[i]].fillna(df[col[i]].mean())
    
    elif df[col[i]].dtype == "float64":
        df[col[i]] = df[col[i]].fillna(df[col[i]].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
n = int(.08 * df.shape[0])

for i in range(len(col)):
    if df[col[i]].nunique() > n :
        df.drop([col[i]] , axis=1 , inplace=True)

```

```python
# AS this column is of date and time and it will be of no use to us

df.drop(['HIC Date'] , axis=1 , inplace=True)
```
